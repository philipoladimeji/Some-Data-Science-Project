```python
# Importing libraries
import pandas as pd
import numpy as np
```

```python
# Read csv file into a pandas dataframe
df = pd.read_csv("train.csv")
```

```python
# Take a look at the first few rows     
df.head(5)
```

```python
##Cleaning of data gear_type column
```

```python
for count in range(len(df)):
    a = len(str(df.loc[count, 'gear_type']))
    if a == 3:
        if df.loc[count, 'Permit_registration_year'] <= 2013:
            df.loc[count, 'gear_type'] = "manuell"
        else :
            df.loc[count, 'gear_type'] = "manuell"
   
        
```

```python
df.loc[22, 'gear_type']
```

```python
##encode the categorical data
from sklearn import preprocessing
lb = preprocessing.LabelBinarizer()
df['gear_type'] = lb.fit_transform(df['gear_type'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
```

```python
tips = sns.load_dataset("tips")
df['gas_type'] = df.dropna(subset=['gas_type'])
sns.catplot(x=df['car_brand'], y=df['gas_type'], data=tips, size = 20);
```
